# Networks are made-up of nodes and ties

* Networks are collections of nodes (also called actors) and ties (also called dyads)

* Nodes can represent individuals, teams, organizations, communities, nations, or artifacts

* Ties are concrete connections linking nodes with respect to a relation of interest

* What keeps nodes and ties together is the concept of relation. As Wasserman and Faust (1994) note: "the set of friendships among pairs of children ina classroom are ties that define a relation

# What are relations about?

Some popular relations investigated in the field of network analytics (Wasserman and Faust, 1994):

* evaluation of one person by another (e.g., friendship, liking, or repsect)

* exchange of material resources (e.g., business transactions lending or borrowing things)

* association or affiliation (e.g., groups of individuals jointly attending an event)

* behavioral interaction (talking together, sending messages)

* movement between places or statuses (e.g., migration or social mobility)

* physical connections (e.g., a road, river, or bridge connecting points)

* formal relations (e.g., authority)

* biological linkages (e.g., kinship or descent)

# Forms of networks

Here are the major forms networks can take:

* one-mode Vs 2-mode (also called bipartite or affiliation networks)

* directed Vs undirected

* weighted Vs unweighted

*Note:* these forms are not mutually exclusive. In fact, they can co-occurr in concrete network instances (e.g., there are networks that are both directed and weighted)

# One-mode networks

Key features:

* One-mode networks involve nodes of the same kind (e.g., individuals or firms but not individuals and firms)

* One mode-networks can be thought as "flat" networks

Example: friendship network among individuals (yellow circles)

# Two-mode networks

Key features:

* Two-mode networks involve nodes of different kind (e.g., individuals and firms, students and modules, users and offerings)

* Two mode-networks can be thought as "multi-layer" networks

* Links among nodes of the same kind are not allowed (students don't attend students, rather they attend modules)

Example: Spotify users (yellow circles) listening to songs in the platform (light blue squares)

# Directed and weighted networks

Both directed and weighted networks can be seen as "feature-rich" networks. Ultimately, they offer tie-level attributes.

* Directed networks are useful to represent flows of information or materials (when it's important to discriminate the source and the recipient) or socially-oriented perceptions (e.g, rivalry) or actions (e.g., followership).

* Weighted networks focus on the intensity of ties.

Example: Think about employees indicating who feel their rivals are and, for each of them, the intensity of the rivalry. Directed, weighted network of rivalry.
